```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
%matplotlib inline
f = h5py.File("test_f.h5", "r")
mu = f['MU'][:]

for it in f:
    print(it)
f.close()

def gJ(n,M):
    p = np.pi/(M + 1)
    return ((M - n + 1) *np.cos(p * n) + np.sin(p*n)*np.cos(p)/np.sin(p))/(M+1)

en = np.linspace(-0.1,0.1,1000)
T = np.array([1 + 0*en, en])
dos = 0.5 * mu[0] * T[0] *gJ(0,mu.size) + mu[1]*T[1]*gJ(1,mu.size)
for i in range(2,mu.size):
    T[i%2] = 2* en * T[(i - 1)%2] - T[i%2]
    dos = dos + mu[i] * T[i%2]*gJ(i,mu.size)
    
f = h5py.File("test_broad.h5", "r")
mu = f['MU'][:]
T = np.array([1 + 0*en, en])
dosS = 0.5 * mu[0] * T[0] *gJ(0,mu.size) + mu[1]*T[1]*gJ(1,mu.size)
for i in range(2,mu.size):
    T[i%2] = 2* en * T[(i - 1)%2] - T[i%2]
    dosS = dosS + mu[i] * T[i%2]*gJ(i,mu.size)
    
f = h5py.File("exact.h5", "r")
mu = f['MU'][:]
T = np.array([1 + 0*en, en])
dosE = 0.5 * mu[0] * T[0] *gJ(0,mu.size) + mu[1]*T[1]*gJ(1,mu.size)
for i in range(2,mu.size):
    T[i%2] = 2* en * T[(i - 1)%2] - T[i%2]
    dosE = dosE + mu[i] * T[i%2]*gJ(i,mu.size)
    
    
plt.plot(en,2*dosE/np.pi/np.sqrt(1.-en**2), en,2*dosS/np.pi/np.sqrt(1.-en**2), en,2*dos/np.pi/np.sqrt(1.-en**2))
```
